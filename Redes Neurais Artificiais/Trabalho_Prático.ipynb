# Trabalho Prático

O objetivo do trabalho será colocar em prática o que vocês estão aprendendo na disciplina de Redes Neurais Artificiais.

Nesse trabalho, vocês serão livres para escolherem o problema e o conjunto de dados que quiserem. No entanto, há algumas recomendações e passos (mínimos) que devem ser seguidos.

O trabalho será dividido em duas partes, detalhadas abaixo.

**Grupos:** Os trabalhos podem ser feitos em grupos de até 3 pessoas (do mesmo Polo). Os grupos devem ser os mesmos nas Partes I e II do trabalho.

**Avaliação:** Cada parte do trabalho será avaliada por meio de uma apresentação presencial. Consulte os agendamentos no seu respectivo Polo. Cada apresentação vale 7,5% da nota da disciplina e deverá ter entre 5 e 10 minutos.

## Parte I

Na Parte I do trabalho, vocês devem escolher um problema de classificação ou regressão, apresentar o problema, e conduzir alguns experimentos iniciais.

### Escolha do Conjunto de Dados

Cada grupo poderá escolher o conjunto de dados e o problema de **classificação** ou **regressão** que quiser. Algumas dicas e recomendações:

- Evite conjuntos de dados muito grandes. Vocês podem ter problemas para executar os algoritmos caso não tenham um Computador com a capacidade necessária.
- Evite conjuntos de dados muito pequenos. O trabalho deve ser minimamente desafiador. Além disso, alguns algoritmos não funcionam bem sem uma quantidade adequada de dados.
- Evite conjuntos de dados textuais ou de imagens. O curso terá outras disciplinas específicas para esses tipos de dados.
- Evite conjuntos de dados com dados faltantes.
- No caso de classificação, evite conjuntos de dados com proporções de classes (rótulos) extremamente desbalanceadas. Esse é um problema de aprendizado de máquina desafiador e fora do escopo dessa disciplina.
- **Evite problemas muito fáceis**! A ideia é melhorar os resultados na Parte II, quando aplicarem os novos conceitos que irão aprender ao longo da disciplina.
- Não é proibido, mas evite utilizar o mesmo conjunto de dados já utilizado em outra disciplina do curso.

Algumas sugestões de repositórios e fontes de dados:

- https://www.kaggle.com/datasets
- https://archive.ics.uci.edu/datasets
- https://zenodo.org/
- https://github.com/awesomedata/awesome-public-datasets
- https://datasetsearch.research.google.com/ (serviço Google para busca de datasets)



### Descrição do Conjunto de Dados

Vocês devem apresentar e descrever o conjunto de dados. Entre as perguntas a serem respondidas:
- Qual a tarefa de predição?
- Qual o tamanho do conjunto e dados? Quantos objetos? Quantas características?
- Quais são as características do conjunto de dados?
- Quais características são categóricas? Quais são numéricas?
- Se for um problema de classificação, qual a proporção de cada classe?
- Se for um problema de regressão, qual a distribuição da variável dependente (resposta)?

### Pré-processamento

Aqui, vocês devem descrever e executar as etapas de pré-processamento necessárias para construir um bom modelo para o problema em questão.

É necessário aplicar alguma transformação no conjunto de dados (e.g., padronização ou reescala)?

O conjunto de dados possui variáveis categóricas? É necessário transformá-las em variáveis numéricas?

### Etapa inicial de construção do modelo

Nessa etapa, vocês devem utilizar uma MLP para abordar o problema proposto. Mais especificamente, aqui vocês ficam restritos a utilizar a biblioteca [`scikit-learn`](https://scikit-learn.org/). Assim, se forem abordar um problema de classificação, devem utilizar a classe [``MLPClassifier``](https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html). Se forem abordar um problema de regressão, devem utilizar a classe [``MLPRegressor``](https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPRegressor.html).

Lembre de executar todas as boas práticas de aprendizado de máquina que discutimos. Por exemplo, separar os dados em conjuntos de treinamento e teste.

Apresente uma avaliação dos modelos treinados. Utilize mais do que uma métrica para a avaliação, se possível. Aqui, vocês devem lembrar do que viram nas disciplinas passadas de classificação e regressão.

Nesta etapa do trabalho, **não é necessário realizar seleção de hiperparâmetros**. Vocês podem treinar os modelos com os hiperparâmetros utilizados como padrão pela `scikit-learn` ou treinar e avaliar os modelos com alguns valores de hiperparâmetros distintos.

Também não é necessário se preocupar com conjunto de validação ou regularização. O objetivo da Parte I é construir uma Rede Neural simples para o problema. Na Parte II, vocês irão evoluir a solução.

**Lembrem:** Na apresentação presencial, vocês devem apresentar o trabalho em um Notebook Python (Preferencialmente Google Colab).

## Parte II

Na Parte II, vocês devem evoluir a Parte I do trabalho, colocando em prática todo o conteúdo da disciplina.

Nesse sentido, vocês devem abordar o mesmo problema da Parte I, mas agora utilizando a biblioteca [``Keras``](https://keras.io/).

Além das etapas da Parte I (pré-processamento e preparação dos dados), vocês devem realizar:
- Avaliação do impacto de alguns hiperparâmetros de maneira criteriosa (não é necessário avaliar todos, pois são muitos, mas vocês devem escolher alguns para um estudo de impacto -- e.g., número e tamanho de camadas ocultas, funções de ativação, *batch_size* e *learning_rate*).
- Retenção de um conjunto de validação e monitoramento da função de custo nos conjuntos de treinamento e validação (demonstrando que o modelo não sofre de *underfitting* nem de *overfitting*).
- Aplicação da heurística de *early stopping*.
- Avaliação do impacto de se utilizar ao menos um tipo de regularização (L2 ou Dropout).
- Repetição do processo de treinamento e teste para comparação dos modelos obtidos nas Partes I e II, incluindo a realização de testes de hipótese. Do mais, vocês devem interpretar os resultados e indicar se um modelo teve desempenho superior ao outro.

**Lembrem:** Na apresentação presencial, vocês devem apresentar o trabalho em um Notebook Python (Preferencialmente Google Colab).

**Observação:** se vocês escolherem um problema "muito fácil", já na Parte I terão resultados difíceis de serem superados. Daí a dica de escolher problemas não triviais. Assim, conseguirão demonstrar melhor que assimilaram os conceitos vistos na disciplina.
