# Trabalho Prtico

O objetivo do trabalho ser colocar em prtica o que vocs esto aprendendo na disciplina de Redes Neurais Artificiais.

Nesse trabalho, vocs sero livres para escolherem o problema e o conjunto de dados que quiserem. No entanto, h algumas recomendaes e passos (mnimos) que devem ser seguidos.

O trabalho ser dividido em duas partes, detalhadas abaixo.

**Grupos:** Os trabalhos podem ser feitos em grupos de at 3 pessoas (do mesmo Polo). Os grupos devem ser os mesmos nas Partes I e II do trabalho.

**Avaliao:** Cada parte do trabalho ser avaliada por meio de uma apresentao presencial. Consulte os agendamentos no seu respectivo Polo. Cada apresentao vale 7,5% da nota da disciplina e dever ter entre 5 e 10 minutos.

## Parte I

Na Parte I do trabalho, vocs devem escolher um problema de classificao ou regresso, apresentar o problema, e conduzir alguns experimentos iniciais.

### Escolha do Conjunto de Dados

Cada grupo poder escolher o conjunto de dados e o problema de **classificao** ou **regresso** que quiser. Algumas dicas e recomendaes:

- Evite conjuntos de dados muito grandes. Vocs podem ter problemas para executar os algoritmos caso no tenham um Computador com a capacidade necessria.
- Evite conjuntos de dados muito pequenos. O trabalho deve ser minimamente desafiador. Alm disso, alguns algoritmos no funcionam bem sem uma quantidade adequada de dados.
- Evite conjuntos de dados textuais ou de imagens. O curso ter outras disciplinas especficas para esses tipos de dados.
- Evite conjuntos de dados com dados faltantes.
- No caso de classificao, evite conjuntos de dados com propores de classes (rtulos) extremamente desbalanceadas. Esse  um problema de aprendizado de mquina desafiador e fora do escopo dessa disciplina.
- **Evite problemas muito fceis**! A ideia  melhorar os resultados na Parte II, quando aplicarem os novos conceitos que iro aprender ao longo da disciplina.
- No  proibido, mas evite utilizar o mesmo conjunto de dados j utilizado em outra disciplina do curso.

Algumas sugestes de repositrios e fontes de dados:

- https://www.kaggle.com/datasets
- https://archive.ics.uci.edu/datasets
- https://zenodo.org/
- https://github.com/awesomedata/awesome-public-datasets
- https://datasetsearch.research.google.com/ (servio Google para busca de datasets)



### Descrio do Conjunto de Dados

Vocs devem apresentar e descrever o conjunto de dados. Entre as perguntas a serem respondidas:
- Qual a tarefa de predio?
- Qual o tamanho do conjunto e dados? Quantos objetos? Quantas caractersticas?
- Quais so as caractersticas do conjunto de dados?
- Quais caractersticas so categricas? Quais so numricas?
- Se for um problema de classificao, qual a proporo de cada classe?
- Se for um problema de regresso, qual a distribuio da varivel dependente (resposta)?

### Etapa inicial de construo do modelo

Nessa etapa, vocs devem utilizar uma MLP para abordar o problema proposto. Mais especificamente, aqui vocs ficam restritos a utilizar a biblioteca [`scikit-learn`](https://scikit-learn.org/). Assim, se forem abordar um problema de classificao, devem utilizar a classe [``MLPClassifier``](https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html). Se forem abordar um problema de regresso, devem utilizar a classe [``MLPRegressor``](https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPRegressor.html).

Lembre de executar todas as boas prticas de aprendizado de mquina que discutimos. Por exemplo, separar os dados em conjuntos de treinamento e teste.

Apresente uma avaliao dos modelos treinados. Utilize mais do que uma mtrica para a avaliao, se possvel. Aqui, vocs devem lembrar do que viram nas disciplinas passadas de classificao e regresso.

Nesta etapa do trabalho, **no  necessrio realizar seleo de hiperparmetros**. Vocs podem treinar os modelos com os hiperparmetros utilizados como padro pela `scikit-learn` ou treinar e avaliar os modelos com alguns valores de hiperparmetros distintos.

Tambm no  necessrio se preocupar com conjunto de validao ou regularizao. O objetivo da Parte I  construir uma Rede Neural simples para o problema. Na Parte II, vocs iro evoluir a soluo.

**Lembrem:** Na apresentao presencial, vocs devem apresentar o trabalho em um Notebook Python (Preferencialmente Google Colab).

## Parte II

Na Parte II, vocs devem evoluir a Parte I do trabalho, colocando em prtica todo o contedo da disciplina.

Nesse sentido, vocs devem abordar o mesmo problema da Parte I, mas agora utilizando a biblioteca [``Keras``](https://keras.io/).

Alm das etapas da Parte I (pr-processamento e preparao dos dados), vocs devem realizar:
- Avaliao do impacto de alguns hiperparmetros de maneira criteriosa (no  necessrio avaliar todos, pois so muitos, mas vocs devem escolher alguns para um estudo de impacto -- e.g., nmero e tamanho de camadas ocultas, funes de ativao, *batch_size* e *learning_rate*).
- Reteno de um conjunto de validao e monitoramento da funo de custo nos conjuntos de treinamento e validao (demonstrando que o modelo no sofre de *underfitting* nem de *overfitting*).
- Aplicao da heurstica de *early stopping*.
- Avaliao do impacto de se utilizar ao menos um tipo de regularizao (L2 ou Dropout).
- Repetio do processo de treinamento e teste para comparao dos modelos obtidos nas Partes I e II, incluindo a realizao de testes de hiptese. Do mais, vocs devem interpretar os resultados e indicar se um modelo teve desempenho superior ao outro.

**Lembrem:** Na apresentao presencial, vocs devem apresentar o trabalho em um Notebook Python (Preferencialmente Google Colab).

**Observao:** se vocs escolherem um problema "muito fcil", j na Parte I tero resultados difceis de serem superados. Da a dica de escolher problemas no triviais. Assim, conseguiro demonstrar melhor que assimilaram os conceitos vistos na disciplina.
